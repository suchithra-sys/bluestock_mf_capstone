```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

nav = pd.read_csv(r"C:\Users\suchithra\Downloads\MutualFundProject\data\processed\02_nav_history_clean.csv")

transactions = pd.read_csv(r"C:\Users\suchithra\Downloads\MutualFundProject\data\raw\08_investor_transactions.csv")

performance = pd.read_csv(r"C:\Users\suchithra\Downloads\MutualFundProject\data\raw\07_scheme_performance.csv")

aum = pd.read_csv(r"C:\Users\suchithra\Downloads\MutualFundProject\data\raw\03_aum_by_fund_house.csv")
```

```python
nav.groupby("date")["nav"].mean().plot(figsize=(10,5))
plt.title("Average NAV Trend")
plt.show()
```

```python
aum.plot(kind="bar")
plt.title("AUM Growth")
plt.show()
```

```python
print(transactions.columns)
```

```python
transactions["transaction_date"] = pd.to_datetime(
    transactions["transaction_date"]
)

sip_trend = transactions.groupby("transaction_date")["amount_inr"].sum()

plt.figure(figsize=(12,6))
sip_trend.plot()

plt.title("SIP Inflow Trend")
plt.xlabel("Transaction Date")
plt.ylabel("Total Amount (INR)")
plt.grid(True)

plt.show()
```

```python
sns.heatmap(nav.corr(numeric_only=True))
plt.show()
```

```python
transactions["age_group"].value_counts().plot(kind="bar")

plt.title("Investor Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

```python
print(transactions.columns.tolist())
```

```python
transactions.groupby("state")["amount_inr"].sum().plot(kind="bar", figsize=(12,6))

plt.title("State-wise Investment Amount")
plt.xlabel("State")
plt.ylabel("Total Investment Amount (INR)")
plt.xticks(rotation=45)

plt.show()
```

```python
# Gender Distribution
transactions["gender"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

```python
# City Tier Distribution
transactions["city_tier"].value_counts().plot(kind="bar")
plt.title("City Tier Distribution")
plt.show()
```

```python
# Payment Mode
transactions["payment_mode"].value_counts().plot(kind="bar")
plt.title("Payment Mode Analysis")
plt.show()
```

```python
# KYC Status
transactions["kyc_status"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("KYC Status Distribution")
plt.ylabel("")
plt.show()
```

```python
sns.heatmap(performance.corr(numeric_only=True), annot=True)
plt.show()
```

```python
performance.groupby("category")["returns"].sum().plot(kind="pie")
plt.show()
```

```python
print(performance.columns.tolist())
```

```python
performance.groupby("category")["return_3yr_pct"].mean().sort_values(ascending=False).plot(
    kind="bar", figsize=(12,6)
)

plt.title("Average 3-Year Return by Category")
plt.xlabel("Category")
plt.ylabel("Return (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
top_funds = performance.sort_values(
    by="return_5yr_pct",
    ascending=False
).head(10)

plt.figure(figsize=(12,6))
plt.bar(top_funds["scheme_name"], top_funds["return_5yr_pct"])
plt.title("Top 10 Funds by 5-Year Return")
plt.xticks(rotation=90)
plt.ylabel("Return (%)")
plt.show()
```

```python
performance["expense_ratio_pct"].hist(bins=20)

plt.title("Expense Ratio Distribution")
plt.xlabel("Expense Ratio (%)")
plt.ylabel("Count")
plt.show()
```

```python
performance["category"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(8,8)
)

plt.title("Fund Category Distribution")
plt.ylabel("")
plt.show()
```

```python
performance["risk_grade"].value_counts().plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Risk Grade Distribution")
plt.xlabel("Risk Grade")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    performance["alpha"],
    performance["sharpe_ratio"]
)

plt.xlabel("Alpha")
plt.ylabel("Sharpe Ratio")
plt.title("Alpha vs Sharpe Ratio")
plt.show()
```

```python
performance.groupby("category")["aum_crore"].sum().sort_values(
    ascending=False
).plot(
    kind="bar",
    figsize=(12,6)
)

plt.title("AUM by Fund Category")
plt.ylabel("AUM (Crore)")
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

cols = [
    "return_1yr_pct",
    "return_3yr_pct",
    "return_5yr_pct",
    "alpha",
    "beta",
    "sharpe_ratio",
    "expense_ratio_pct"
]

plt.figure(figsize=(10,8))

sns.heatmap(
    performance[cols].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Fund Metrics Correlation")
plt.show()
```
